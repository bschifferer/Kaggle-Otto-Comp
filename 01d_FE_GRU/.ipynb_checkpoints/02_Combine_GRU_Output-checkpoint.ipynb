```python
import glob
import cudf
import gc
import pickle
```

```python
igfold = 0
```

```python
files = glob.glob(
    './data_folds/fold_' + str(igfold) + '/data_gru/gru_pred_xgb_*.parquet'
)
print(files)
print(len(files))
```

```python
os.system('mkdir -p ./data_folds/fold_' + str(igfold) + '/candidates/train/gru_scores_20' + '/')
for ifile, file in enumerate(files):
    if ifile==0:
        df = cudf.read_parquet(file)
        pre = file.split('/')[-1].split('_')[3][0]
        ty = file.split('/')[-1].split('_')[4][0:2]
        print(pre, ty)
        df = df.rename(columns={'gru_scores': 'gru_' + str(pre) + '_' + str(ty)})
    else:
        df_tmp = cudf.read_parquet(file)
        pre = file.split('/')[-1].split('_')[3][0]
        ty = file.split('/')[-1].split('_')[4][0:2]
        print(pre, ty)
        df_tmp = df_tmp.rename(columns={'gru_scores': 'gru_' + str(pre) + '_' + str(ty)})
        df = df.merge(
            df_tmp,
            how='outer',
            on=['session', 'cand']
        )
        df = df.fillna(0.0)
        del df_tmp
        gc.collect()

df['session'] = df['session'].astype('int32')
df['cand'] = df['cand'].astype('int32')
dtypes = list(df.dtypes.reset_index().values)
for dtype in dtypes:
    if dtype[1].name == 'float64':
        df[dtype[0]] = df[dtype[0]].astype('float32')
df.to_parquet(
    './data_folds/fold_' + str(igfold) + '/candidates/train/gru_scores_20' + '/cand.parquet'
)
```

```python
files = glob.glob(
    './data_folds/fold_' + str(igfold) + '/data_gru/gru_pred_sub_*.parquet'
)
print(files)
```

```python
from tqdm import tqdm

sub_sessions = pickle.load(open('./data/split/sub_sessoins_10.pickle', 'rb'))
os.system('mkdir -p ./data_folds/fold_' + str(igfold) + '/split/candidates/sub/gru_scores_20')
for i, session_list in enumerate(sub_sessions):
    for ifile, file in enumerate(files):
        if ifile==0:
            df = cudf.read_parquet(file)
            df = df[df['session'].isin(session_list)]
            pre = file.split('/')[-1].split('_')[3][0]
            ty = file.split('/')[-1].split('_')[4][0:2]
            print(pre, ty)
            df = df.rename(columns={'gru_scores': 'gru_' + str(pre) + '_' + str(ty)})
        else:
            df_tmp = cudf.read_parquet(file)
            df_tmp = df_tmp[df_tmp['session'].isin(session_list)]
            pre = file.split('/')[-1].split('_')[3][0]
            ty = file.split('/')[-1].split('_')[4][0:2]
            print(pre, ty)
            df_tmp = df_tmp.rename(columns={'gru_scores': 'gru_' + str(pre) + '_' + str(ty)})
            df = df.merge(
                df_tmp,
                how='outer',
                on=['session', 'cand']
            )
            df = df.fillna(0.0)
            del df_tmp
            gc.collect()
    df['session'] = df['session'].astype('int32')
    df['cand'] = df['cand'].astype('int32')
    dtypes = list(df.dtypes.reset_index().values)
    for dtype in dtypes:
        if dtype[1].name == 'float64':
            df[dtype[0]] = df[dtype[0]].astype('float32')
    df.to_parquet(
        './data_folds/fold_' + str(igfold) + '/split/candidates/sub/gru_scores_20' + '/cand_' + str(i) + '.parquet'
    )
```
